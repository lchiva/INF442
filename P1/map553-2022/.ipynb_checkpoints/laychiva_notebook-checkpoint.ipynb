```python
# Encode soil type ordinally
def categorical_encoding_soil(input_df):
    data = input_df.copy()
    data['Soil_Type'] = 0
    for i in range(1,41):
        data['Soil_Type'] += i*data[f'Soil_Type{i}']
    return data
```

```python
# Encode soil type
train_data = categorical_encoding_soil(train_data)
```

```python
# Encode soil type ordinally
def categorical_encoding_wilderness(input_df):
    data = input_df.copy()
    data['Wilderness_Type'] = 0
    for i in range(1,5):
        data['Wilderness_Type'] += i*data[f'Wilderness_Area{i}']
    return data
```

```python
train_data = categorical_encoding_wilderness(train_data)
```
